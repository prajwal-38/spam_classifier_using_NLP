```python
import pandas as pd
df = pd.read_csv('spam.csv')
df = df[['v1','v2']]
df[['label','messages']] = df[['v1','v2']]
df = df.drop(['v1','v2'], axis=1)
df.head()
```

```python
import re
import nltk
nltk.download('stopwords') 
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
stop = set(stopwords.words('english'))
for i in range(len(df)):
    ne = re.sub('[^A-Za-z]', ' ', df['messages'][i])
    ne = ne.lower()
    ne = ne.split()
    ne1 = []
    for word in ne:
        if word not in stop:
            ne1.append(word)
    
    singles = [ps.stem(plural) for plural in ne1]
    singles = ' '.join(singles)
    corpus.append(singles)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
x = cv.fit_transform(corpus).toarray()
x
```

```python
cv.vocabulary_
```

```python
y = pd.get_dummies(df['label']).astype(int)
```

```python
y = y.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect = MultinomialNB().fit(xtrain,ytrain)
```

```python
ypreds = spam_detect.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(ytest,ypreds)
```

```python
print(classification_report(ytest,ypreds))
```
